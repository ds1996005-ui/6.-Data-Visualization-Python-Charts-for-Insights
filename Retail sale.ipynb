```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean
df = pd.read_csv(r"C:\Users\HP\Downloads\test.csv")
df.columns = df.columns.str.strip()  # remove extra spaces
```

```python
# Group by Neighborhood
top_neighborhoods = df.groupby("Neighborhood")["SalePrice"].sum().nlargest(10)

# Plot
plt.figure(figsize=(10,6))
top_neighborhoods.plot(kind="bar", color="teal")
plt.title("Top 10 Neighborhoods by Total Sale Price")
plt.xlabel("Neighborhood")
plt.ylabel("Total Sale Price")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Insights 
print("Insights:")
print("1. The tallest bar shows the neighborhood with the highest overall sales volume, signaling market dominance.")
print("2. Mid-ranked neighborhoods contribute significantly, suggesting a balanced distribution of demand across multiple areas.")
print("3. The lowest bar among the top 10 still represents strong activity, highlighting that even secondary neighborhoods play a role in overall revenue.")
```

```python
# Create datetime column
df["SaleMonth"] = pd.to_datetime(df["YrSold"].astype(str) + "-" + df["MoSold"].astype(str), format="%Y-%m")

# Group and count sales
monthly_sales = df.groupby("SaleMonth").size()

# Plot
plt.figure(figsize=(12,6))
monthly_sales.plot(marker="o", color="blue")
plt.title("Monthly House Sales Trend")
plt.xlabel("Month")
plt.ylabel("Number of Sales")
plt.grid(True)
plt.tight_layout()
plt.show()

# After plotting the line chart
print("Insights:")
print("1. Clear seasonal peaks appear mid-year, suggesting stronger buyer activity in summer months.")
print("2. Noticeable dips occur in winter, reflecting lower transaction volumes during colder seasons.")
print("3. A sharp spike in [specific month/year] indicates unusual demand, possibly tied to market or policy changes.")
```

```python
# Histogram of Sale Price Distribution
plt.figure(figsize=(10,6))
plt.hist(df["SalePrice"], bins=30, color="orange", edgecolor="black")
plt.title("Distribution of Sale Prices")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# After plotting the histogram
print("Insights:")
print("1. The distribution is right-skewed, meaning most homes sell at moderate prices while a few luxury properties push the tail higher.")
print("2. The bulk of transactions cluster between $100k$200k, showing the dominant market segment.")
print("3. Outliers above $500k highlight premium neighborhoods or custom-built homes.")
```

```python
# Scatter Plot of Living Area vs Sale Price
plt.figure(figsize=(10,6))
plt.scatter(df["GrLivArea"], df["SalePrice"], alpha=0.6, color="purple")
plt.title("GrLivArea vs SalePrice")
plt.xlabel("Above Ground Living Area (sq ft)")
plt.ylabel("Sale Price")
plt.grid(True)
plt.tight_layout()
plt.show()

# After plotting the histogram
print("Insights:")
print("1. The distribution is right-skewed, meaning most homes sell at moderate prices while a few luxury properties push the tail higher.")
print("2. The bulk of transactions cluster between $100k$200k, showing the dominant market segment.")
print("3. Outliers above $500k highlight premium neighborhoods or custom-built homes.")
```
